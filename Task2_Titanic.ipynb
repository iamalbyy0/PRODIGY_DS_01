```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_df = pd.read_csv(r"C:\Users\albin\Downloads\titanic\test.csv")
test_df.head(5)
```

```python
train_df = pd.read_csv(r"C:\Users\albin\Downloads\titanic\train.csv")
train_df.head(5)
```

```python
test_df.isnull().sum()
```

```python
train_df.isnull().sum()
```

```python
# Drop 'Cabin' column (too many missing values)
train_df.drop('Cabin', axis=1, inplace=True)
```

```python
# Fill missing 'Age' with median
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna(train_df['Embarked'].mode()[0])
```

```python
train_df.isnull().sum()
```

## Step 5: Exploratory Data Analysis (EDA)

```python
# Count plot of Survived
sns.countplot(x='Survived', data=train_df)
plt.show()
```

```python
# Count plot of Survived by Sex
sns.countplot(x='Survived', hue='Sex', data=train_df)
plt.show()
```

```python
# Count plot of Survived by Pclass
sns.countplot(x='Survived', hue='Pclass', data=train_df)
plt.show()
```

```python
# Age distribution
sns.histplot(train_df['Age'], bins=30, kde=True)
plt.show()
```

```python
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

## SUMMARY
The Titanic dataset was first cleaned by fixing missing values and removing unnecessary columns. Through exploratory data analysis, clear patterns were found, such as higher survival rates among women and first-class passengers. Age and ticket fare also showed a noticeable impact on survival, helping to better understand the factors that influenced passenger outcomes.
